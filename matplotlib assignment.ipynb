```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

Note: Also add title, xlabel, and ylabel to the plot.
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
A scatter plot is a type of graph that displays the values of two variables in a set of data using dots using Cartesian coordinates.
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('random data')
plt.show()
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
The subplot() function creates multiple subplot layouts in a single figure. With it, I can create common subplot layouts, including enclosed figures.
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3= np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2 ,2 ,1)
plt.plot(x1,y1)

plt.subplot(2 ,2 ,2)
plt.plot(x2,y2)

plt.subplot(2 ,2 ,3)
plt.plot(x3,y3)

plt.subplot(2 ,2 ,4)
plt.plot(x4,y4)

plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
A Bar Plot is a bar chart, a graphical representation of data using rectangular bars. The length of bar corresponds to the value it represents. They are useful when working with
dicrete or categorical data
```

```python
plt.bar(company,profit)
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()
```

```python
plt.barh(company,profit)
plt.title('Company Profits')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```

```python
A box plot is graphical representation that displays the distribution and statistical information of dataset. The are useful when we want to understand the overall
distribution of dataset.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.title('Box_Plot')
plt.boxplot([box1, box2])
plt.show()
```
